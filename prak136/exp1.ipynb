```python
import matplotlib.pyplot as plt
import numpy as np
import uncertainties as unp
from scipy.optimize import curve_fit
```

```python
x = [1, 2, 3, -1, -2, -3]
siny = [0.10947113555609, 0.21894996563084, 0.328492391072299, -0.109466303659085, -0.218916763373329, -0.328317914107691]

x = np.array(x)
y = np.array(siny)
print(len(x), len(y))
```

```python
def f(x, a, b):
    return x*a + b
```

```python
popt, perr = curve_fit(f, x, y)
print(popt, np.sqrt(np.diag(perr)))

plt.rcParams["figure.figsize"] = (12,7)
plt.tick_params(which='major', direction='in', width=1.5, length=8)
plt.tick_params(which='minor', direction='in', width=1.5, length=4)
plt.minorticks_on()
plt.grid(which='major', linewidth=0.5, linestyle='dashed', color='black')

plt.errorbar(x, y, fmt='s')
plt.plot(x, f(x, *popt))
# plt.ylim(0, 0.4)
plt.xlabel(r'$m$', fontsize="large")
plt.ylabel(r'$sin \phi$', fontsize="large")
plt.title(r'Синуса угла дифракции от порядка для зеленого цвета')
plt.savefig("sin_phi_zel.png")
plt.show()
```
