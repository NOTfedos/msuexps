```python
def f1(x, a):
    return a*x

def f2(x, a, b):
    return a*x + b
```

```python
popt, perr = curve_fit(f1, np.array(a_obr_arr), np.array(dx_arr))
print(popt, np.sqrt(np.diag(perr)))

plt.rcParams["figure.figsize"] = (12,7)
plt.tick_params(which='major', direction='in', width=1.5, length=8)
plt.tick_params(which='minor', direction='in', width=1.5, length=4)
plt.minorticks_on()
plt.grid(which='major', linewidth=0.5, linestyle='dashed', color='black')

plt.errorbar(a_obr_arr, dx_arr, xerr=a_obr_arr_err, yerr=dx_arr_err, fmt='s')
plt.plot(a_obr_arr, f1(np.array(a_obr_arr), *popt))
plt.xlabel(r'$\frac{1}{a}$, $\frac{1}{м}$', fontsize="large")
plt.ylabel(r'$\Delta x$, мм', fontsize="large")
plt.title(r'Зависимость ширины полосы $\Delta x$ от $\frac{1}{a}$')
plt.savefig("dx(1_a).png")
plt.show()
```

```python
popt, perr = curve_fit(f2, np.array(b_a_arr), np.array(dx_arr))
print(popt, np.sqrt(np.diag(perr)))

plt.rcParams["figure.figsize"] = (12,7)
plt.tick_params(which='major', direction='in', width=1.5, length=8)
plt.tick_params(which='minor', direction='in', width=1.5, length=4)
plt.minorticks_on()
plt.grid(which='major', linewidth=0.5, linestyle='dashed', color='black')

plt.errorbar(b_a_arr, dx_arr, xerr=b_a_arr_err, yerr=dx_arr_err, fmt='s')
plt.plot(b_a_arr, f2(np.array(b_a_arr), *popt))
plt.xlabel(r'$\frac{b}{a}$', fontsize="large")
plt.ylabel(r'$\Delta x$, мм', fontsize="large")
plt.title(r'Зависимость ширины полосы $\Delta x$ от $\frac{b}{a}$')
plt.savefig("dx(b_a).png")
plt.show()
```

```python
popt, perr = curve_fit(f1, np.array(a_arr), np.array(d_arr))
print(popt, np.sqrt(np.diag(perr)))

plt.rcParams["figure.figsize"] = (12,7)
plt.tick_params(which='major', direction='in', width=1.5, length=8)
plt.tick_params(which='minor', direction='in', width=1.5, length=4)
plt.minorticks_on()
plt.grid(which='major', linewidth=0.5, linestyle='dashed', color='black')

plt.errorbar(a_arr, d_arr, xerr=a_arr_err, yerr=d_arr_err, fmt='s')
plt.plot(a_arr, f1(np.array(a_arr), *popt))
plt.xlabel(r'$a$, мм', fontsize="large")
plt.ylabel(r'$d$, мм', fontsize="large")
plt.title(r'Зависимость расстояния между источниками $d$ от расстояния $a$ источников до бипризмы')
plt.savefig("d(a).png")
plt.show()
```
