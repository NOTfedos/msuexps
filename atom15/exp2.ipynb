```python
%matplotlib widget
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
import os


I = np.array([330, 377, 380, 414], dtype=float)
dk = np.array([0.194,
0.2205,
0.223,
0.242]) * 1e5 / 4.67
dk_err = [0.01, 0.01, 0.01, 0.01]

def func(x, a, b):
    return a*x + b

popt, perr = curve_fit(func, I, dk)
print(popt, np.sqrt(np.diag(perr)))

plt.errorbar(I, dk, xerr=[1]*4, yerr=dk_err, fmt='s')
plt.plot(I, func(I, popt[0], popt[1]))
plt.xlabel(r'I, мА')
plt.ylabel(r'H, Эрстед')
plt.title(r'Зависимость H от I для красной линии Zn')
plt.savefig(os.path.join(".", "ims", "Caliber_H.png"))
plt.show()
```
