```python
import matplotlib.pyplot as plt
import numpy as np
import uncertainties as unp
from scipy.optimize import curve_fit
```

```python
def f1(x, a):
    return a*x

def f2(x, a, b):
    return a*x + b
```

```python
m = np.array([1, 2, 3, 4, 5, 6])
r = np.array([135, 186, 229, 259, 292, 311])

r = r * 0.01 / 3

r = r ** 2

yerr = [0.03 for x in r]
```

```python
popt, perr = curve_fit(f2, m, r)
print(popt, np.sqrt(np.diag(perr)))

plt.rcParams["figure.figsize"] = (12,7)
plt.tick_params(which='major', direction='in', width=1.5, length=8)
plt.tick_params(which='minor', direction='in', width=1.5, length=4)
plt.minorticks_on()
plt.grid(which='major', linewidth=0.5, linestyle='dashed', color='black')

plt.errorbar(m, r, yerr=yerr, fmt='s')
plt.plot(m, f2(m, *popt))
plt.xlabel(r'$m$', fontsize="large")
plt.ylabel(r'$r^2$, $мм^2$', fontsize="large")
# plt.xlim(0, 15)
# plt.ylim(0, 60)
plt.title(r'Зависимость квадрата радиуса $r^2$ кольца от порядка $m$')
plt.savefig("zel_r_m.png")
plt.show()
```

```python
r_k = np.array([7, 99, 169, 220, 262, 295])
r_zh = np.array([6, 83, 161, 205, 247, 282])
r_s = np.array([6, 98, 155, 198, 230, 256])

m = np.array([0, 1, 2, 3, 4, 5])

r_k = (r_k * 0.01 / 3) ** 2
r_zh = (r_zh * 0.01 / 3) ** 2
r_s = (r_s * 0.01 / 3) ** 2
```

```python
plt.rcParams["figure.figsize"] = (12,7)
plt.tick_params(which='major', direction='in', width=1.5, length=8)
plt.tick_params(which='minor', direction='in', width=1.5, length=4)
plt.minorticks_on()
plt.grid(which='major', linewidth=0.5, linestyle='dashed', color='black')


popt, perr = curve_fit(f2, m, r_k)
print("Красный", popt, np.sqrt(np.diag(perr)))
plt.errorbar(m, r_k, yerr=yerr, fmt='s', label="Красный")
plt.plot(m, f2(m, *popt))


popt, perr = curve_fit(f2, m, r_zh)
print("Желтый", popt, np.sqrt(np.diag(perr)))
plt.errorbar(m, r_zh, yerr=yerr, fmt='s', label="Желтый")
plt.plot(m, f2(m, *popt))


popt, perr = curve_fit(f2, m, r_s)
print("Красный", popt, np.sqrt(np.diag(perr)))
plt.errorbar(m, r_s, yerr=yerr, fmt='s', label="Синий")
plt.plot(m, f2(m, *popt))

plt.xlabel(r'$m$', fontsize="large")
plt.ylabel(r'$r^2$, $мм^2$', fontsize="large")
# plt.xlim(0, 15)
# plt.ylim(0, 60)
plt.title(r'Зависимость квадрата радиуса $r^2$ кольца от порядка $m$ для различных светодиодов')
plt.legend()
plt.savefig("r_m.png")
plt.show()
```
